```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
cols = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'car_name']
```

```python
file = "./auto-mpg-data/auto-mpg.csv"
mpg = pd.read_csv(file, names = cols, na_values = '?')
```

```python
mpg.head()
```

```python
def strip_car_name(row):
    return row.car_name.strip()[1:-1]
```

```python
mpg['car_name'] = mpg.apply(strip_car_name, axis = 1)
mpg.head()
```

```python
sns.pairplot(data = mpg)
```

```python
sns.lmplot(x = 'displacement', y = 'horsepower', hue = 'origin', data = mpg)
```

```python
sns.boxplot(y = 'mpg', x = 'origin', data = mpg)
```

```python
sns.distplot(mpg.mpg)
```

```python
sns.violinplot(x = 'origin', y = 'mpg', data = mpg)
```
